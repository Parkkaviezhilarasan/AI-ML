Predicting the students mark based on the study hours, sleep hours and the previous exam marks using linear regression

```python
import pandas as pd #to_handle_dataset
import numpy as np #numerical_operation

data=pd.DataFrame(columns=['Study_Hours','Sleep_Hours','Previous_Exam_Marks', 'Final_marks'])
data.loc[0]=[2,6,45,50 ]
data.loc[1]=[4,7,55,65]
data.loc[2]=[6,8,60,75]
data.loc[3]=[1,5,40,42]
data.loc[4]=[5,7,65,78]
data.loc[5]=[3,6,50,58]
data.loc[6]=[7,8,70,85]
print(data)

#Seperating the X and Y

X=data[["Study_Hours","Sleep_Hours","Previous_Exam_Marks"]]
Y=data["Final_marks"]

print(X.shape)
print(Y.shape)
```

TRAINNING

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X,Y)
print(model.coef_)
print(model.intercept_)
```

```python
y_pred=model.predict(X)
print(list(Y))
print(list(y_pred))
```

```python
from sklearn.metrics import mean_squared_error
rmse=np.sqrt(mean_squared_error(Y,y_pred))
print(rmse)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y,y_pred)
print(r2)
```

```python
new=np.array([[5,7,60]])
predict_mark=model.predict(new)
print(predict_mark)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3,3))
plt.scatter(data["Study_Hours"], data ["Final_marks"])
plt.xlabel("Study_Hours")
plt.ylabel("Final_marks")
plt.title("Study_Hours vs Final_marks")
plt.show()

plt.figure(figsize=(3,3))
plt.scatter(Y,y_pred)
plt.xlabel("Final_marks")
plt.ylabel("y_pred")
plt.title("Final_marks vs y_pred")
plt.show()
```
